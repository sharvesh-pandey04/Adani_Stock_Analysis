```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("ADANIENT.csv",header=2)
```

```python
df.rename(columns={df.columns[0]:'Date'},inplace=True)
```

```python
df['Date']=pd.to_datetime(df['Date'],errors='coerce')
```

```python
df=df.dropna(subset=['Date'])
```

```python
df.columns=['Dates','Close','High','Low','Open','volume']
```

```python

df['Dates'] = pd.to_datetime(df['Dates'])

```

```python
df=df.sort_values(by='Dates')
```

```python
print("Max Price:",df['Close'].max())
```

```python
df['Daily_Return']=df['Close'].pct_change()
```

```python
df[['Dates','Close','Daily_Return']].head()
```

```python
df['MA_10']=df['Close'].rolling(10).mean()
df['MA_20']=df['Close'].rolling(20).mean()
df[['Close','MA_10','MA_20']].head()
```

```python
recent=df[df['Dates']>'2023-01-01']
recent.head(20)
```

```python
high_return=df[df['Daily_Return']>0.02]
high_return[['Dates','Daily_Return']]
```

```python
df['Month']=df['Dates'].dt.month
monthly_avg=df.groupby('Month')['Close'].mean()
monthly_avg
```

```python
df.to_csv("final_output.csv",index=False)
```
